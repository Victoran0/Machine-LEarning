#### Distribution Plots

##### Distribution Plot

```python
# sns.distplot(crash_df['not_distracted'])

# To remove the graph estimation of our dataset, we can use the arg kde=False
# sns.distplot(crash_df['not_distracted'], kde=False)

# Bins is like a bucket that divides our data up btw different intervals, and it determines how many bars we are going to see accross our data
sns.displot(crash_df['not_distracted'], kde=True, bins=25)

plt.show()
```

##### Joint Plot

```python
 # A joint plot is going to be used to compare two distribution and plot a scatter plot by default

# To generate a regression line, we use the reg val for kind arg
# sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

# KDE means kernel density estimation
# sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')

# Using the hex kind
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')

plt.show()
```

##### Pair Plots

```python
# A pair plot is going to plot the relationships across the entire data frame numerical values
# sns.pairplot(crash_df)

tips_df = sns.load_dataset('tips')

# With the hue arg, we can pass in categorical data and the charts will be colorized based on that categorical data
# THe palette is for styling
sns.pairplot(tips_df, hue='sex', palette='Blues')

plt.show()
```

##### Rug Plot

```python
# A rugplot is going to plot a single column of data points in a data frame as sticks 
# and with a rugplot, we will see a more dense number of lines where the amount is most common, like a histogram is taller where values are but for a rugplot, it is denser where the vals are more common
sns.rugplot(tips_df['tip'])
```

#### Categorical Plots

##### Bar Plots

```python
# sns.barplot(x='sex', y='total_bill', data=tips_df)

# The default aggregation of the data is based off their mean but we can change that to median, std, cov, and we can make our own functions
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.cov)
```

##### Count Plot

```python
# A count plot is similar to a bar plot but the estimator is just going to count the number of occurences
sns.countplot(x='sex', data=tips_df)
```

##### Box Plot

```python
# A box plot is going to allow us to compare different variables and show us the quartiles of the data
# The bar tha goes across the box is the median and the box is going to extend one standard deviation from our median
# The whiskers extend to other data aside what is in our std aside from those points that are considered outliers (circle at the tops)
# hue gives us a chance to pass in additional category to plot oncv 
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.legend(bbox_to_anchor=(1.01, 1.0), loc='upper left')
```

##### Violin Plot

```python
# It is a combination of a box plot as well as a kde plot
# WHile a box plot is going to correspond to data points, a violin plot uses the kde estimation of our data point to create its plots
# sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')

# We have an arg split, if we set this to true, it will allow us to compare how the category compare to each other
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

##### Strip Plot

```python
plt.figure(figsize=(8,5))
# A strip plot draws a scatter plot repreenting all the different data points where one variable is categorical
# we are often going to see them used to show different observations with a boxplot that is going to represent hte average distributions
# Setting jitter True spread our data out a little bit
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex')
```

##### Swarm Plot

```python
# A swarm plot is like a strip plot but points are going to be adjusted so that they do not overlap, it looks like the combination btw a violin and a strip plot
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

#### Matrix Plots

##### Heatmaps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

# To create a heatmap with data, we are basically going to try to setup in a matrix format variablex
# For example, in our crash_df, to create a heatmpa, we are not to align data as a row with their indexes, we are to set their column values to become the rows
# There are two ways to get this done

# one way is to call correlation function on our df and basically corr is going to tell us how influencial a variable is going to be on our result, so that is going to create that
crash_mx = crash_df.corr(numeric_only=True)

# annot arg is the numbers in the center of each part of our heatmap
sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
# We can also create a matrix plot by using a pivot table
# Lets test with a new dataset
flights = sns.load_dataset('flights')
# A pivot table is another way to get our rows put into columns and then we correlate the data
flights = flights.pivot_table(index='month', columns='year', values='passengers')
# We then create an heatmap from this data, the linecolor would not show up unless we define the line width
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1)
```
