```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
bars = plt.bar(labels, values)

# We can also add titles etc using the same commands we used in line graph.

# To set hatch for a particular bar we use the below function which sets the hatch for the first bar:
# bars[0].set_hatch('/')

# We can use circle as our hatch value, for the second we use the O letter:
# bars[1].set_hatch('O')

# To use star, we use *
# bars[2].set_hatch('*')

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())

# We can try resizing it
plt.figure(figsize=(6,4))
plt.show()
```
