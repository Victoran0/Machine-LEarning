#### Gas Prices in Different Countries (in USD)

```python
# Plotting OUr graph based on a csv file data
gas = pd.read_csv('./data/gas_prices.csv')

plt.figure(figsize=(18,10))
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight': 'bold', 'fontsize': 22})

# plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
# plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
# plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')
# # The below for .year is best for multi words data
# plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

countries_to_look_at = ['USA', 'Australia', 'Canada', 'South Korea']
for country in gas.columns[1:]:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
# Moving the legend outside of the graph
plt.legend(bbox_to_anchor=(1.01, 1.0), loc='upper left')
# Using the list iterator [], we know start index comes first, then the end index then the skip ratio, i.e[::3] all start, all end and skip by 3
plt.xticks(gas.Year[::3])

# To add another year past the list above, we can use
plt.xticks(gas.Year[::3].tolist()+[2011])

plt.savefig('Gas Price_Figure.png', dpi=400)

plt.show()
```

#### Histogram: Fifa Player Stats


```python
fifa = pd.read_csv('./data/fifa_data.csv')

# fifa.head(5)
```

```python
# To build a histogram, we use the plt.hist() syntax
# The bins arg give the x axis vals following its numbering
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.yticks([0,100])
plt.show()
```

#### Pie Charts

```python
# Pie Charts work best by passing in a list of numbers
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]

labels = ['left', 'right']
colors = ['#abcdef', '#00ffff']
plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Foot Preference of Fifa Players')

plt.show()
```

#### Pie Chart 

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

# We can use the ggplot style(read doc for more info-Matplotlib style Options)
plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa[fifa.Weight >= 200].count().iloc[0]

# The explode splits the graph apart so it does not touch
explode = (.4,.1,0,0, .4)

plt.title('Weight Distribution of FIFA Players (in lbs)')

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
plt.pie(weights, labels = labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)


plt.show()
```

#### Whiskers / Box Plot

```python
# Getting the Overall score of all the players from barcelona
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']
plt.style.use('default')
plt.figure(figsize=(5,8))
# To use facecolor arg, we have to set the patch_artist to True
# The median props sets the styles etc for the median line
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth': 2})

# Using other style setting options
colors
for box in boxes['boxes']:
    # Set edge color
    box.set(color='#4286f4', linewidth=2)

    # Change fill color
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating') 
plt.show()
```
