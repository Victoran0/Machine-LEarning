#### Load Neccessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph
###### We can search through the documentation https://matplotlib.org/stable/api/pyplot_summary.html with ctrl+f

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# To plot our graph, it is x,y arg in the .plot function
# We can add a label/formula/logic the plot of our graph
# plt.plot(x, y, label='2x', color='#04f0f0', linewidth=2, marker='.',linestyle='--', markersize=10, markeredgecolor='blue')

# Resize our Graph
# The figsize is the value of the x and y axis, the dpi specifies pixels per inch we want our graph to take
plt.figure(figsize=(5,3), dpi=100)

# To give a title to our graph, we use the below:
plt.title('Our First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
# Using shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'r^--',label='2x')

# Adding a label to x and y
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Line number Two
# When we want to handle values involved with decimals, we can use
# For the arange function, the args are (start num, limit num, diff between each val)
x2 = np.arange(0,4,0.5)
# plt.plot(x2, x2**2, 'b', label='x^2')
# If we have a graph stopped and we want to show its future projection, we use:
# WE plot the first 4
plt.plot(x2[:5], x2[:5]**2, 'b', label='x^2')
# For the last values:
plt.plot(x2[4:], x2[4:]**2, 'b--')



# Changing the range of ticks on the axis
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

# The label for the graph line will not show without the legend function
plt.legend()

# To save our graph, we can use the below function
plt.savefig('mygraph.png', dpi=300)

# The correct way to show the graph is by running plt.show()
plt.show()
```
