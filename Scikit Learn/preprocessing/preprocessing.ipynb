```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('drawndata1.csv')
```

```python
df.head()
```

```python
X = df[['x', 'y']].values
y = df['z'] == 'a'
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

## Without using any scalers, we can see the values of our axis differs and a lot of outliers, this can impact out algorithm alot

## Using the StandardScaler to scale the data to an extent

```python
from sklearn.preprocessing import StandardScaler, QuantileTransformer
X_new = StandardScaler().fit_transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y)
plt.show()
```

## Using the standard scaler, we can see that our axis have now being scaled but we still have the outliers

## Using the quantiletransformer to manae the outliers: We can see the scale on the x and y axis equaling

```python
X_new = QuantileTransformer(n_quantiles=252).fit_transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y)
plt.show()
```

### The difference in the data is that we are transforming and scalling with the quantiles as opposed to the mean and standard deviation
